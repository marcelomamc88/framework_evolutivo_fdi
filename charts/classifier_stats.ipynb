```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
fontsize_xylabel = 10
fontsize_tick = 10
```

## SAM KNN

### Inicial

```python
data = [
    [17.1, 82.9, 0, 0, 0, 0, 0],
    [13.4, 0, 86.6, 0, 0, 0, 0],
    [0, 0, 0, 100, 0, 0, 0],
    [0, 0, 0, 0, 100, 0, 0],
    [0, 80.6, 0, 0, 0, 19.4, 0],
    [0, 82.3, 0, 0, 0, 0, 17.7]
]

inputed = ['Falha 1', 'Falha 2', 'Falha 3', 'Falha 4', 'Normal 1', 'Normal 2']
predicted = ['Rejeição', 'Falha 1', 'Falha 2', 'Falha 3\n(rejeição)', 'Falha 4\n(rejeição)', 'Normal 1\n(rejeição)', 'Normal 2\n(rejeição)']

df = pd.DataFrame(np.array(data)/100, index=inputed, columns=predicted)

fig, axs = plt.subplots(figsize=(10.5,4))
cm = sns.heatmap(df, annot=True, fmt='.1%', cmap='Blues', rasterized=True, annot_kws={"fontsize":fontsize_tick})
cm.set_xticklabels(cm.get_xticklabels(), rotation = 0, size=fontsize_xylabel)
cm.set_yticklabels(cm.get_yticklabels(), rotation = 0, size=fontsize_xylabel)

colorbar = cm.collections[0].colorbar
colorbar.set_ticks([0, .25, .5, .75, 1])
colorbar.set_ticklabels(['0%','25%', '50%', '75%', '100%'])

fig = cm.get_figure()
fig.savefig("6_classifier_samknn_heatmap.pdf", bbox_inches='tight')
```
