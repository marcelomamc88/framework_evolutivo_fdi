```python
import pandas as pd
import numpy as np
from sklearn.metrics import *
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fontsize_xylabel = 13
fontsize_tick = 14
```

## Half Space Tree

```python



data = [
    [0.899,0,0,0.11],
    [0.894,0,0,0.106],
    [0.1486, 0.8414, 0, 0.01],
    [0,0,0.99,0.01],
    [0.1319, 0,0,0.8681],
    [0,0.5029,0,0.4971]
]


df = pd.DataFrame(np.array(data), index=['Normal 1', 'Normal 2', 'Falha 1', 'Falha 2', 'Falha 3', 'Falha 4'], columns=['Normal', 'Falha 1', 'Falha 2', 'Desconhecido'])

fig, axs = plt.subplots(figsize=(9,5))
cm = sns.heatmap(df, annot=True, fmt='.1%', cmap='Blues', rasterized=True, annot_kws={"fontsize":fontsize_tick})
cm.set_xticklabels(cm.get_xticklabels(), rotation = 0, size=fontsize_xylabel)
cm.set_yticklabels(cm.get_yticklabels(), rotation = 0, size=fontsize_xylabel)

plt.xlabel('Predito')
plt.ylabel('Inputado')

plt.xticks(rotation=0, ha="center")

colorbar = cm.collections[0].colorbar
colorbar.set_ticks([0, .25, .5, .75, 1])
colorbar.set_ticklabels(['0%','25%', '50%', '75%', '100%'])

fig = cm.get_figure()
fig.savefig("6_operation_confusion_matriz.pdf", bbox_inches='tight')
```
